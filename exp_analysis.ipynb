# Exploratory Analysis of the Artists Dataset

```python
# special IPython command to prepare the notebook for matplotlib
%matplotlib inline 
from IPython.display import Image
#Array processing
import numpy as np
#Data analysis, wrangling and common exploratory operations
import pandas as pd
from pandas import Series, DataFrame

#For visualization. Matplotlib for basic viz and seaborn for more stylish figures
import matplotlib.pyplot as plt
import seaborn as sns

from pandas.plotting import scatter_matrix
```

```python
#Reading the csv file into the pandas data frame

df_data = pd.read_csv('artists.csv', encoding='latin1')

#Return the first five rows of the dataset
df_data.head()
```

# Task 1: Statistical Exploratory Data Analysis
Let us start with getting to know the dataset. The first task will be to get some basic information by using Pandas features.

```python
#Task 1-a: Print the column names of the df_data data frame
print(">> Task 1-a: Column names in df_data data frame are: \n", list(df_data))

#Task 1-b: Find the number of rows and columns in the df_data data frame.
row_count = df_data.shape[0]
column_count = df_data.shape[1]
print("\n>> Task 1-b: Number of Rows: %s \n Number of Columns: %s\n " % (row_count, column_count))

#Task 1-c: Print the descriptive detail for 'Year of Birth' column of the df_data 
yearofbirth = df_data["year_of_birth"].describe()
print ("\n>>Task 1-c: Descriptive details of 'Year of Birth' column are\n",yearofbirth)

#Task 1-d: Print ALL the unique values of role and Countries  .
role_list  = df_data["role"].unique()
unique_country_list = df_data["country"].unique()
print ("\n>>Task 1-d:\n")
print("Unique Role List\n")
print(role_list)
print("----------------------------------------------------------------------")
print("Unique Country List\n")
print(unique_country_list) 
```

# Task 2: Aggregation & Filtering & Rank
In this task, we will perform some aggregation and filtering operations. 
Then, we will apply ranking on the results for some tasks. 
Pandas has a convenient and powerful syntax for aggregation, filtering, and ranking. 

```python
#Task 2-a: Print the details of pianists and violinists from Brazil 
artists_from_brazil = df_data.query('country=="Brazil" & role=="pianist" | country=="Brazil" & role=="violinist"')
print (">>Task 2-a: The details of 'pianist' and 'violinist' from Brazil are %s" % (artists_from_brazil))

#Task 2-b: Find out the total number of arts made by artists born between 1958 and 1972.
arts = df_data.query('role=="artist" & year_of_birth>1957 & year_of_birth<1973')
arts_made_1958_1972 = len(arts)
print ("\n>>Task 2-b: The total number of arts made by artists born between 1958-1972 were %s" %(arts_made_1958_1972))

#Task 2-c: Find out the top 5 countries with roles 'artist' and 'guitarist'.
top5_artist = df_data.loc[(df_data['role'] == "artist")]['country'].value_counts().head() #value_counts() returns counts of unique values
top5_guitarist= df_data.loc[(df_data['role'] == "guitarist")]['country'].value_counts().head()
print ("\n>>Task 2-c:\nThe top 5 countries with artists are  \n%s" % (top5_artist))
print ("\nThe top 5 countries with guitarists are \n%s" % (top5_guitarist))
```

# Task 3: Visualization
In this task, we will perform a number of visualization tasks to get some intuition about the data.

```python
sns.set_style('whitegrid')
sns.set(font_scale = 1.3)

#Task 3-a: Plot the count for each role. 
plt.figure(figsize=(25,10))
sns.countplot(x="role",data=df_data)
plt.title("Count of each role")

#Task 3-b: Draw a pie chart that represents each role.
df = df_data['role'].value_counts().values
labels = df_data['role'].value_counts().index
plt.figure(figsize=(10,10))
plt.pie(df,labels=labels ,autopct='%1.0f%%', pctdistance=1.1, labeldistance=1.2)
plt.title("Pie Chart representing Role")
plt.show()
```
